### Research Question 1 Result:

Even though there are some outliers in each of four different income groups, we can still see the the birth rate goes down as the income increase. However, there is no correlation relationship between them. It's just trends. I am assuming that most of people can have higher income in the developed countries. The more developed the country, the better the relative welfare, the less the burden of supporting the old, so that even if they have only a few children they can maintain a normal life, those parents have no need to have so many children. The income of developing countries is relatively low compared with that of developed countries, and their social welfare is not as good as that of developed countries, so they have to have many children to maintain their future life.

## The relationship between internet users and birth rate in High Income Group countries

```python
%matplotlib inline
import matplotlib.pyplot as plt

fig, axis = plt.subplots()
# Grid lines, Xticks, Xlabel, Ylabel

axis.yaxis.grid(True)
axis.set_title('Relationship between internet users and birth rate in High Income Group Countries',fontsize=10)
axis.set_xlabel('High Income Group Birth_Rate',fontsize=10)
axis.set_ylabel('High Income Group Internet_users',fontsize=10)

X = data_highIncome['Birth rate']
Y = data_highIncome['Internet users']

axis.scatter(X, Y)
plt.show()
```

```python
np.corrcoef(data_highIncome['Birth rate'],data_highIncome['Internet users'])
```

As the scatter plot shows, in the high income group countries, there is Correlation relationship between internet users and birth rate. As the birth rate goes up, the internet users fall down.

## Now let's combine all four scatter plots into one chart

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
fig, axis = plt.subplots()
# Grid lines, Xticks, Xlabel, Ylabel

axis.yaxis.grid(True)
axis.set_title('Relationship between internet users and birth rate in different Income Group Countries',fontsize=10)
axis.set_xlabel('Birth_Rate',fontsize=10)
axis.set_ylabel('Internet_users',fontsize=10)

axis.scatter(X, Y, label = 'High')
axis.scatter(data_Upper['Birth rate'], data_Upper['Internet users'], label = 'Upper_Middle')
axis.scatter(data_lower['Birth rate'], data_lower['Internet users'], label = 'Lower_Middle')
axis.scatter(data_low['Birth rate'], data_low['Internet users'], label = 'Low')

plt.legend(loc = 0, mode = "expand", ncol = 4)
plt.show()
```

## Research Question 2 Result:

As the combined chart shows above, it's clear to see that as the birth rate goes up, the numbers of internet users fall down.
